# **Dissimilarity Matrix generation for Binary attributes**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('/content/student-mat.csv')
data.head()
dataBinary=data[['sex','schoolsup','famsup','paid','activities','nursery','higher','internet','romantic']]
dataBinary.head()
```

```python
dataBinary=dataBinary.replace('no',0)
dataBinary=dataBinary.replace('yes',1)
dataBinary.head()
```

```python
n=np.array(dataBinary[['schoolsup','famsup']])
n=n.reshape(-1,2)
n.shape
```

```python
m=np.array(dataBinary[['romantic','internet']])
m=m.reshape(-1,2)
m.shape
```

```python
from scipy.spatial import distance
dist_matrixB=distance.cdist(n,m)
dist_matrixB.shape
sns.heatmap(dist_matrixB)
plt.show()
```

# **Dissimilarity Matrix for Numerical attributes**

```python
dataNumeric=data[['age','Medu','Fedu','traveltime','studytime','failures']]
dataNumeric.head()
```

```python
n1=np.array(dataNumeric[['age','failures']])
n1=n1.reshape(-1,2)
n1.shape
```

```python
m1=np.array(dataNumeric[['Medu','Fedu']])
m1=m1.reshape(-1,2)
m1.shape
```

```python
from scipy.spatial import distance
dist_matrixN=distance.cdist(n1,m1)
dist_matrixN.shape
sns.heatmap(dist_matrixN,cmap='YlGn')
plt.show
```

# **Dissimilarity Matrix for Nominal Attributes**

```python
dataNominal=data[['Mjob','Fjob','reason','guardian']]
dataNominal.replace('at_home','home')
dataNominal=(dataNominal.astype('category'))
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
dataNominal['guardian']=lb.fit_transform(dataNominal['guardian'])
dataNominal['Mjob']=lb.fit_transform(dataNominal['Mjob'])
dataNominal['Fjob']=lb.fit_transform(dataNominal['Fjob'])
dataNominal['reason']=lb.fit_transform(dataNominal['reason'])
dataNominal.head()
```

```python
n2=np.array(dataNominal)
n2.reshape(-1,2)
n2.shape
```

```python
m2=np.array(dataNominal)
m2.reshape(-1,2)
m2.shape
```

```python
from scipy.spatial import distance
dist_matrixNominal = distance.cdist(n2,m2)
sns.heatmap(dist_matrixNominal,cmap='PiYG')
plt.show
```
